# Fish Kills Across Florida

The Florida red tide refers to the process by which high concentrations of the Algae Karenia brevis become present, creating toxins that have immense consequences for animal and human well being. Each year, Florida witnesses massive fish killings across the coast resulting in millions of lost dollars for Florida's fishing and tourism industries. Even more important, the red tide poses a great danger to human health as the toxins produced can cause serious respiratory illnesses to people who inhale them.

In this notebook I will explore the fish kills that have occured from the year 2000 to 2019, due to algae blooms and the red tide, and use this data to forecast future fish kills.

## Visualize data

First I will import relevant libraries and glance at the data.

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import Roboto_Condensed_Bold
import datetime

# Data
data = pd.read_csv("data/FishKillResultReport.csv")
fish_kill = data[data['Call Category'] == 'Fish Kill']
display(fish_kill.head())
print("Shape of data: {}".format(fish_kill.shape))
```

## Stationarize Data

Time series models assume stationarity(constant mean, constant variance, autocovariance does not depend on time). One way to test whether the data is stationary is through the Dickey-Fuller test which measures how strongly a time series is defined by a trend.

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(df, ts):
    """
    Test stationarity using moving average statistics and Dickey-Fuller test
    Source: https://www.analyticsvidhya.com/blog/2016/02/time-series-forecasting-codes-python/
    """
    
    # Determing rolling statistics
    rolmean = df[ts].rolling(window = 12, center = False).mean()
    rolstd = df[ts].rolling(window = 12, center = False).std()
    
    # Plot rolling statistics:
    orig = plt.plot(df[ts], 
                    color = 'blue', 
                    label = 'Original')
    mean = plt.plot(rolmean, 
                    color = 'red', 
                    label = 'Rolling Mean')
    std = plt.plot(rolstd, 
                   label = 'Rolling Std')
    plt.legend(loc = 'best')
    plt.title('Rolling Mean & Standard Deviation for %s' %(ts))
    plt.xticks(rotation = 45)
    plt.show(block = False)
    plt.close()
    
    # Perform Dickey-Fuller test:
    # Null Hypothesis (H_0): time series is not stationary
    # Alternate Hypothesis (H_1): time series is stationary
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(df[ts], 
                      autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], 
                         index = ['Test Statistic',
                                  'p-value',
                                  '# Lags Used',
                                  'Number of Observations Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(df = new_data.resample('M').sum(), ts = 'Fish Kills')
```

Based on the p-value (0.000023) from the augmented Dickey-Fuller test we can conclude that the data is stationary (does not have a time-dependent structure). This is important because we now know that the time series data can be used with statistical modeling and machine learning methods. If the p-value was greater than 0.05 then we would need to correct for stationarity using techniques such as tranforming, smoothing, differencing, polynomial fitting and decomposition.

## Plot ACF/PACF

The parameters for an ARIMA model are as follows:
- p (Number of Auto Regressive terms) Seasonality
- d (Number of Integrated or Difference terms) Trend
- q (Number of Moving Average Terms) Noise

In order to find the parameters to feed into the ARIMA model we can use ACF (Autocorrelation function) and PACF (Partial Autocorrelation Function) plots. ACF shows the correlation between the current observations and the observations at previous time spots. PACF still measures the correlation between two observations but also considers the effect that other observations may have on those two observations.

```python
def plot_acf_pacf(df, ts):
    """
    Using Statsmodels to plot:
    1. auto-correlation function (ACF)
    2. partial auto-correlation (PACF) plots
    """
    from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
    
    fig, ax = plt.subplots(1,2,figsize=(10,5))

    # Plot ACF
    plot_acf(df, title = "ACF {}".format(ts), lags = 40, ax = ax[0]);
    # Plot PCF
    plot_pacf(df, title = "PACF {}".format(ts), lags = 40, ax = ax[1]);
    
    plt.tight_layout()
    plt.show()
```

```python
# plot ACF and PACF
plot_acf_pacf(df = new_data.resample('M').sum(), ts = 'Fish Kills')
```

The above two plots show the Auto correlation function and partial auto correlation function for the time series data. The points that are above the gray area in the graph are considered as points that are statistically significant. Since both the ACF and PACF plots have only three statistically significant points, both the ACF and PACF follow a geometric series and therefore the ARMA model would be the best fit.

## Test Model and Forecast- ARIMA

ARIMA stands for Autoregressive Integrated Moving Average. Since ARIMA takes in p, d, and q, the first thing I will do is use grid search to find the optimal set of parameters for our model.

```python
def find_param(df, ts):
    
    import itertools
    p = range(0,2)
    d = range(0,2)
    q = range(0,2)
    pdq = list(itertools.product(p,d,q))
    
    from statsmodels.tsa.arima_model import ARIMA

    for choice in pdq:
        
        mod = ARIMA(np.float64(df[ts]), order= choice)
        results_ = mod.fit(disp = -1)

        #Calculate RSS and RMSE
        len_results = len(results_.fittedvalues)
        ts_modified = df[ts][-len_results:]
        rss = sum((results_.fittedvalues - ts_modified)**2)
        rmse = np.sqrt(rss/ len(df[ts]))

        print("ARIMA{} - RSS: {} | RMSE: {}".format(choice, rss, rmse ))
```

```python
print("The range of the data is from {} fish kills to {} fish kills".format(
      new_data.resample("M").sum().min()[0], new_data.resample('M').sum().max()[0]))
find_param(df = new_data.resample('M').sum(), ts = 'Fish Kills')
```

Based on the grid search we can see that the parameters p = 1, d = 0, and q = 1 produced the lowest root mean squared error and therefore are the parameters that should be used when building the ARIMA model. An RMSE of 93.43 isn't too bad considering the range of values from 9 to 1069 in the dataset. Next I'll build and plot the ARIMA model

```python
def build_arima(df, ts, p, d, q):
    from statsmodels.tsa.arima_model import ARIMA
    mod = ARIMA(np.float64(df[ts]), order= (p,d,q))
    results_ = mod.fit(disp = -1)
        
    #Calculate RSS and RMSE
    len_results = len(results_.fittedvalues)
    ts_modified = df[ts][-len_results:]
    rss = sum((results_.fittedvalues - ts_modified)**2)
    rmse = np.sqrt(rss/ len(df[ts]))
    
    #Plot graph with real and predicted values
    plt.figure(figsize = (12,4))
    plt.plot(df[ts], label = 'Actual')
    plt.plot(df.index, results_.fittedvalues, color = 'blue', label = 'ARIMA')
    plt.title("ARIMA model with param: p={}, d={}, q={}".format(p,d,q))
    plt.legend()
    plt.show()
```

```python
build_arima(df = new_data.resample("M").sum(), ts = "Fish Kills", p = 1, d =0, q = 1)
```

Now that I've plotted the ARIMA model on top of the actual values, its time to do some forecasting! Since the data points are on a monthly basis I'll forecast 36 months out or 3 years into the future. Additionally I'm curious to see how accounting for seasonlaity (d = 1) effects the models forecast, so I'll plot two graphs: one with seasonlity and one without.

```python
def ARIMA_forecast(df, ts, p, d, q):
    
    # Create ARIMA and get forecasts
    from statsmodels.tsa.arima_model import ARIMA
    mod = ARIMA(np.float64(df[ts]), order= (p,d,q))
    results = mod.fit(disp = -1)
    pred_uc = results.forecast(steps = 36)
    
    # Confidence intervals
    conf_bottom = []
    conf_top = []
    for val in pred_uc[2]:
        conf_bottom.append(val[0])
        conf_top.append(val[1])

    #Dataframe from forecast results
    index = pd.date_range(start="2020-01-31",end="2022-12-31", freq = 'M')
    data = {'Sample Forecasts': list(pred_uc[0]),
                'Standard Error': list(pred_uc[1]),
                'Confidence_bottom': conf_bottom,
                'Confidence_top': conf_top}
    frame = pd.DataFrame(data = data, index = index)

    # Plot
    ax = df[ts].plot(label='observed', figsize=(14, 7))
    frame['Sample Forecasts'].plot(ax=ax, label='Forecast', color = 'blue')
    ax.fill_between(frame.index,
                    frame['Confidence_bottom'],
                    frame['Confidence_top'], color='dimgrey', alpha=.25)
    ax.set_xlabel('Date')
    ax.set_ylabel('Fish Kills')
    plt.legend()
    plt.title("ARIMA 3 Year Forecast | param: p={}, d={}, q={}".format(p,d,q))
    plt.show()
```

```python
# Forecast without considering seasonality
ARIMA_forecast(df = new_data.resample('M').sum(), ts = 'Fish Kills', p = 1, d = 0, q=1)
# Forecast with consideration of seasonality
ARIMA_forecast(df = new_data.resample('M').sum(), ts = 'Fish Kills', p = 1, d = 1, q=1)
```

As we can see from the two graphs above, accounting for seasonality greatly effects the three year forecast. The top graph, not accounting for seasonality, projects that reported fish kills will remain at the average which is promising news. On the other hand, the bottom graph which accounts for seasonality projects that the reported fish kills will increase in the next three years which is very dissapointing news.
